```python
import random
import numpy as np
import datetime
```

```python
data1 = [[random.randint(0,99) for j in range (1000)] for i in range(1000)]
count = 0
start_time = datetime.datetime.now()
for i in range(1000):
    for j in range(1000):
        if data1[i][j] > 50:
            count +=1
end_time = datetime.datetime.now()
data1_time = end_time - start_time
print(count)
print('Using time',data1_time.microseconds)
```

record computing time

```python
start_time =datetime.datetime.now()
data2 = np.array(data1)
count = np.sum(data2 >50)
end_time = datetime.datetime.now()
data2_time = end_time - start_time
print(count)
print('Using time', data2_time.microseconds)
```

```python
ratio_time = data2_time/data1_time
print(ratio_time)
```
