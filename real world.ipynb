ตรวจสอบว่าจุดที่สุ่มมาอยู่ในวงกลมหรือไม่ ใช้วิธีการของ Monte caro
วฺิธีการของวงกลม 1 หน่วย 
หาระยะทางแบบยูคลิด d^2 = x^2 + y^2

```python
#สร้างจุดที่มีค่าระหว่าง -1 ถึง 1
#สุ่มจำนวน 10000 , 100000 ,1000000
total_point = 10000
point_in_circle = 0
point_out_circle = 0
for i in range(total_point):
    x = random.random()*2-1
    y = random.random()*2-1
    d = math.sqrt(x**2+ y**2)
    if d > 1:
        print(d,"is not in circle")
        point_out_circle +=1
    else:
        print(d,"is in circle")
        point_in_circle +=1
        
print('Total point in the circle',point_in_circle)
print('Total point out the circle',point_out_circle)
print('The approximation of pi is',(point_in_circle*4)/total_point)
#ค่าประมาณของ pi = จุดในวง*4/จน.จุดทั้งหมด
```

Risk management

m = จำนวนโครงการ
n = จำนวนวิธีการประเมิณ
นำผลการประเมิณมารวมเพื่อดูค่าผลประโยชน์
คัดเลือกว่าจะทำโครงการอะไร
ดูว่าแบบประเมิณไหนให้ผลประโยชน์มากสุด
หาค่า optimization

```python
#เลือกทำโครงการใด
#ให้ผลประโยชน์เท่าไหร่
#[โครงการ],[threshold]
projects = [[[35,34,25,36,48],10],
            [[25,26,38,14,32],15],
            [[28,12,36,25,28],-10]]
threshold = [26,27,37,26,30]

n_project = len(projects)
# print(n_project)

n_policy = len(threshold)
# print(n_policy)
ret = [0]*n_policy #[0,0,0,0,0]

for i in range(n_project):
    for j in range(n_policy):
        if projects[i][0][j] > threshold[j]:
            ret[j] += projects[i][1]
            
max_index = 0
for i in range(1,len(ret)):
    if ret[i] > ret[max_index]:
        max_index = i
        
print('Evaluation method',max_index+1)
print('Expected return' , ret[max_index])
    
```

```python
#cleqn code using chatGPT
import numpy as np

projects = [
    [35, 25, 28],
    [34, 26, 12],
    [25, 38, 36],
    [36, 14, 25],
    [48, 32, 28]
]

thresholds = [26, 27, 37, 26, 30]
benefits = [10, 15, -10]

def calculate_benefit(project, threshold, benefits):
    scores = [benefits[i] if project[i] >= threshold else 0 for i in range(len(project))]
    return sum(scores)

sum_benefits = [calculate_benefit(project, thresholds[i], benefits) for i, project in enumerate(projects)]

for i, project in enumerate(projects):
    print(f'Project {i+1} {project} has a score of {sum_benefits[i]}')

best_expected_return = np.max(sum_benefits)
best_project_index = sum_benefits.index(best_expected_return) + 1
print(f'The maximum expected return is {best_expected_return}')
print(f'The best project is Project {best_project_index}')
```
